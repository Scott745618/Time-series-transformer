```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.interpolate import CubicSpline
```

```python
path = r'C:\Users\SGrah\OneDrive\Documents\Scott\Masters\Thesis\Data\GDP data Sep-21.csv'

df = pd.read_csv(path, header=0, usecols=[0,1,2,3,4,5,6,7], nrows=112, index_col=[0])
df
```

```python
# Step 1: Create monthly index
df.index = pd.to_datetime(df.index, format="%b-%y")
dates_m = pd.date_range(df.index.min(), df.index.max(), freq="M")

# Step 2: Map quarterly index to integer positions
x_q = np.arange(len(df))  # quarterly positions
x_m = np.linspace(0, len(df)-1, len(dates_m))  # monthly positions

# Step 3: Fit cubic spline for ALL columns (matrix)
cs = CubicSpline(x_q, df.values, axis=0)

# Step 4: Evaluate spline at monthly positions
values_m = cs(x_m)

# Step 5: Build new DataFrame with same column names
df_m = pd.DataFrame(values_m, index=dates_m, columns=df.columns)

print("\nMonthly disaggregated data:")
print(df_m.head(12))
```
